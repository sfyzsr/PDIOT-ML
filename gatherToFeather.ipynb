```python
import numpy as np
import pandas as pd
import os
```

```python
respeck = pd.DataFrame()
thingy = pd.DataFrame()
count = 0
for dirname in os.listdir("2022"):
    for filename in os.listdir("./2022/"+dirname):
        path = "./2022/"+dirname+"/"+filename
        if(filename[0]=="R"):
            df = pd.read_csv(path)
            respeck = pd.concat([respeck,df])
        if(filename[0]=="T"):
            df = pd.read_csv(path)
            thingy = pd.concat([thingy,df])
        count=count+1
        if(count%200==0):
            print(count)
    
respeck.shape
```

```python
respeck.drop(columns=['sensor_type','notes','recording_id'], axis=1, inplace=True)
```
