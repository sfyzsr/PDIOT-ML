```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
import random

from tensorflow.python.client import device_lib
print(device_lib.list_local_devices())
```

```python
import pandas as pd

data = pd.read_csv("./Respeck_s1732873_Climbing stairs_04-10-2022_14-29-16.csv")
data
```

```python
# concat all csv to one
import os
# for dirname in os.listdir("2022"):
#     for filename in os.listdir("./2022/"+dirname):
#         print(filename)
i=0
j=0
column_names = {"timestamp": float(),
"accel_x": float(),
"accel_y": float(),
"accel_z": float(),
"gyro_x": float(),
"gyro_y": float(),
"gyro_z": float(),
"sensor_type":"",
"activity_type":"",
"activity_code":"",
"subject_id":"",
"notes":"",
"recording_id":""}
respeck = pd.DataFrame(columns=column_names)
for filename in os.listdir("./2022/s1732873/"):
    # print(filename)
    path = "./2022/s1732873/"
    if(filename[0]=="R"):
        df = pd.read_csv(path+filename)
        respeck = pd.concat([respeck,df])
    if(filename[0]=="T"):
        j=j+1
respeck
```

```python
# label the motion to digits

activites = respeck.loc[0,"activity_type"]
labelDf=pd.DataFrame({'activity_type':activites, 'label':range(len(activites))})
new = pd.merge(respeck, labelDf, how='inner', on = 'activity_type')
new.reset_index()
new.drop(columns=['sensor_type', 'activity_code','subject_id','notes','recording_id'], axis=1, inplace=True)
new['label'] = pd.to_numeric(new['label'])
# store in feather
new.to_feather("respeck.feather")
new
```

```python
r = pd.read_feather("respeck.feather")

batchSize = 32

labels = r.pop('label')
labels = labels.values.astype(float)

r.drop(columns=['timestamp','activity_type'], axis=1, inplace=True)
r = r.values.astype(float)

dataset = tf.data.Dataset.from_tensor_slices((r, labels))
for feat, targ in dataset.take(5):
  print ('Features: {}, Target: {}'.format(feat, targ))
  # print(feat.shape)
train_dataset = dataset.shuffle(len(r)).batch(32)
print(train_dataset)
```

```python
def get_compiled_model():
  model = tf.keras.Sequential([
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(10, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
  ])

  model.compile(optimizer='adam',
                loss='binary_crossentropy',
                metrics=['accuracy'])
  return model
with tf.device("gpu:0"):
  model = get_compiled_model()
  model.fit(train_dataset, epochs=2)
  model.save("my_model")
```

```python
reconstructed_model = keras.models.load_model("my_model")
print(type(reconstructed_model))
```

```python
test_accuracy = tf.keras.metrics.Accuracy()
for (x, y) in train_dataset:
    # training=False is needed only if there are layers with different
    # behavior during training versus inference (e.g. Dropout).
    # print(x.batch(1))
    # logits = model(x, training=False)
    logits = reconstructed_model(x, training=False)
    prediction = tf.argmax(logits, axis=1, output_type=tf.int32)
    test_accuracy(prediction, y)
print("Test set accuracy: {:.3%}".format(test_accuracy.result()))
```
